# Exploratory Data Analysis

## Load data

```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
iris = datasets.load_iris()
```

## Build a DataFrame

```python
X = iris.data
Y = iris.target
df = pd.DataFrame(X, columns = iris.feature_names)
```

```python
print(df.head()) # first five rows
```

```python
print(df.info())
```

```python
print(df.describe())
```

## Visual EDA

```python
from pandas.plotting import scatter_matrix
```

```python
_ = scatter_matrix(df, 
                      c = Y,            # color by species
                      figsize = [8, 8], # size of figures
                      s = 150,          # marker size
                      marker = 'D')     # marker shape
```

# scikit-learn to fit a classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 6)
knn.fit(iris['data'], iris['target']) # features as a NumPy array
                                      # target as a NumPy array
```
